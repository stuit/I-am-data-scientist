```python
import numpy as np  

import seaborn as sns
```

```python
np.random.seed(2)
```

Сгенерируем выборку

```python
def linear_func(x):
    target_k = 2
    target_b = 9
    return target_k * x + target_b

def linear_func2(x):
    target_k = 5
    target_b = -2
    return target_k * x + target_b
```

```python
xes = np.linspace(0, 1, 11)

print(xes)
```

```python
target_ys = np.array([linear_func(x) for x in xes])
print(target_ys)
```

```python
noise = np.random.normal(0, 0.1, 11)
noisy_target_ys = target_ys + noise

print(noisy_target_ys)
```

```python
sns.lineplot(x = xes, y = target_ys, color = "g")
sns.lineplot(x = xes, y = noisy_target_ys, color = "b")
```

Линейная регрессия sklearn

```python
from sklearn.linear_model import LinearRegression

X = xes.reshape((-1, 1))

reg = LinearRegression().fit(X, target_ys)

print("Linear coefs:", reg.coef_, reg.intercept_)
```

```python
reg = LinearRegression().fit(X, noisy_target_ys)

print("Linear coefs:", reg.coef_, reg.intercept_)
```

```python
regression_predict = reg.predict(X)

sns.lineplot(x = xes, y = target_ys, color = "g")
sns.lineplot(x = xes, y = noisy_target_ys, color = "b")
sns.lineplot(x = xes, y = regression_predict, color = "r")

```

```python
class OwnLinearRegression:
    def __init__(self, k = 5, b = -2):
        self.k = k
        self.b = b

    def predict(self, X):
        return [x * self.k + self.b for x in X]
```

```python
lin_reg = OwnLinearRegression()

linear_regression_predict = lin_reg.predict(np.array([xes]))

print(linear_regression_predict)
print(target_ys)
```
