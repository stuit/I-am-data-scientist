```python
import cv2     # для работы с видео
import math   # для математических вычислений
import matplotlib.pyplot as plt    
%matplotlib inline
```

```python
count = 0
videoFile = "Tom and jerry.mp4"
cap = cv2.VideoCapture(videoFile)   # загрузка видео 
frameRate = cap.get(5) # частота кадров
x=1
while(cap.isOpened()):
  frameId = cap.get(1) # номер текущего кадра
  ret, frame = cap.read()
  if (ret != True):
      break
  elif (frameId % math.floor(frameRate) == 0):
    	filename ="frame%d.jpg" % count;count+=1
    	cv2.imwrite(filename, frame)

cap.release()
```
